# Laboratory 3 : Linear Combination and Vector Spaces

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

$$
A=\begin{bmatrix} 2 \\ 4\end{bmatrix} \\
B=\begin{bmatrix} 3 \\ -5 \end{bmatrix} \\
ScatterA = Rangelist \cdot \begin {bmatrix}  2 \\ 4  \end{bmatrix},  
ScatterB = Rangelist \cdot \begin {bmatrix}  2 \\ 4  \end{bmatrix}
$$

```python
vectA = np.array([2,4])
vectB = np.array([3,-5])

Rangelist= np.arange(-20,20,0.25)

ScatterA = plt.scatter(Rangelist*vectA[0],Rangelist*vectA[1], label="ScatterA")
ScatterB = plt.scatter(Rangelist*vectB[0],Rangelist*vectB[1], label="ScatterB")

plt.xlim(-20,20)
plt.ylim(-20,20)
plt.axhline(y=0, color='green', alpha = .5)
plt.axvline(x=0, color='green')
plt.grid()
plt.legend()
plt.show()
```

$$
C=\begin{bmatrix} 3 \\ 4\end{bmatrix} \\
D=\begin{bmatrix} 5 \\ -6\end{bmatrix} \\
Span = \left\{
    \begin{array}\\
        X \cdot \begin {bmatrix} 3 \\ 4\end{bmatrix},
        Y \cdot \begin {bmatrix} 5 \\ -6\end{bmatrix} \\
    \end{array}
\right\} \\
$$

```python
SpanC = np.array([3,4])
SpanD = np.array([5,-6])

Rangelist2 = np.arange(-40,40,5)

X, Y = np.meshgrid(Rangelist2,Rangelist2)

SpanT = SpanC + SpanD
spanTx = X*SpanC[0] + Y*SpanD[0]
spanTy = X*SpanC[1] + Y*SpanD[1]
plt.scatter(spanTx,spanTy, s=5, alpha=0.75)

plt.xlim(-400,400)
plt.ylim(-400,400)
plt.axhline(y=0, color='Red')
plt.axvline(x=0, color='Red')
plt.grid()
plt.show()
```
