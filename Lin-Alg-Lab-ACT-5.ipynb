```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Task 1

$$
A = \begin{bmatrix} 8 \ 9\\3 \ 5 \\\end{bmatrix}\\ or \\
A = \left\{ \begin{array}\\ 8x + 9y \\ 3x + 5y \end{array} \right. 
$$

```python
A = np.array([
    [8, 9],
    [3, 5]
])

plt.xlim(0,20)
plt.ylim(0,20)
plt.title("2D Vector")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.quiver(0,0, A[0,0], A[0,1], angles='xy', 
           scale_units='xy',scale=1, color='green', label = "First Rowspace") 
plt.quiver(0,0, A[1,0], A[1,1], angles='xy', 
           scale_units='xy',scale=1, color='black', label = "Second Rowspace")
plt.grid()
plt.legend()
plt.show()
```

```python
A = np.array([
    [8, 9],
    [3, 5]
])

plt.xlim(0,20)
plt.ylim(0,20)
plt.title("2D Vector")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

c = np.arange(0,20)

plt.plot(c*A[0,0], c*A[0,1], color='green', label = "First rowspace")
plt.plot(c*A[1,0], c*A[1,1], color='black', label = "First rowspace") 


plt.grid()
plt.legend(loc ="upper left")
plt.show()
```

## Task 2

$$
B = \begin{bmatrix} 1 \ 9 \ 5 \\\end{bmatrix}\\
$$
$$
B = \left\{ \begin{array}\\ x + 9y + 5z\\ \end{array} \right. 
$$

$$
B2 = \begin{bmatrix} 4 \ 2 \ 5 \\\end{bmatrix}\\
$$
$$
B2 = \left\{ \begin{array}\\ 3x + 2y + 5z\end{array} \right. 
$$

```python
B = np.array([1,9,5])
B2 = np.array([3,2,5])
fig = plt.figure()
ax1 = fig.gca(projection='3d')

ax1.set_xlim([0, 10])
ax1.set_ylim([0, 10])
ax1.set_zlim([0, 10])
c = np.arange(0,10)

ax1.quiver(0,0,0, B[0], B[1], B[2], 
           arrow_length_ratio=0.2, colors='r', label ="Non-Continous")
ax1.plot(c*B2[0],c*B2[1],c*B2[2],
           color='green', label ="Continous")

plt.legend()
plt.show()
```

$$
C =
\begin{bmatrix}
1 \ 9 \ 5\\
2 \ 5 \ 7 \\
8 \ 3 \ 1 \\
\end{bmatrix}\\
$$

$$
C = \left\{ \begin{array}
\\ x + 9y + 5z 
\\ 2x + 5y + 7z 
\\ 8x + 3y + z \end{array} \right. 
$$

```python
C = np.array([
    [1,9,5],
    [2,5,7],
    [8,3,1]
])

fig = plt.figure()
ax2 = fig.gca(projection='3d')

ax2.set_xlim([0, 15])
ax2.set_ylim([0, 15])
ax2.set_zlim([0, 15])

origin = (0,0,0)

ax2.quiver(origin, origin, origin, C[:,0], C[:,1], C[:,2], 
           arrow_length_ratio=0.2, colors=['red','blue','green'])

plt.show()
```
