### Task 1

```python
X = np.array([1,8,-6,7])

class Solver:
    def __init__(arg):
        arg.squared = [SQ ** 2 for SQ in X]
        arg.norm=(sum([arg.squared[lis] for lis in range(len(arg.squared))]))**(1/2)
        arg.modulo=arg.norm
        arg.rounded = round(arg.modulo,2)
        
def Modulo_find():
    return Solver()

M = Modulo_find

print("The modulo of the given is:",M().modulo)
print("The modulo of the given rounded to 2 decimals is:",M().rounded)
```

### Task 3

$$
Equation = ((A^2 + B^2) \cdot (A * (B + A*B)./C))*||A + B + C||
$$
$$
A = \begin{bmatrix}-0.4\\0.2\\-0.6\end{bmatrix},
B = \begin{bmatrix}-0.2\\0.2\\1\end{bmatrix},
C = \begin{bmatrix}0.2\\-0.1\\-0.5\end{bmatrix}
$$

```python
A = np.array([-0.4,0.2,-0.6])
B = np.array([-0.2,0.2, 1])
C = np.array([0.2,-0.1,-0.5])

FrsEq = np.add(A**2,B**2)
SndEq = np.add(B,np.multiply(A,B))
TrdEq = np.divide(np.multiply(A,SndEq),C)
FrtEq = abs(np.add(A,B))
Answer = np.multiply(np.dot(FrsEq,SndEq),FrtEq)

Answer
```

```python
figure = plt.figure()
out = figure.gca(projection='3d')
out.quiver(0,0,0, Answer[0], Answer[1], Answer[2], color="black")
out.set_xlim([0, 0.5])
out.set_ylim([0, 0.5])
out.set_zlim([0, 0.5])
plt.show()
```
